### Exercise 1: Analyzing NYT click-through rates based on age

```r
data1 <- read.csv(file.path("data/dds_datasets/dds_ch2_nyt/nyt1.csv"))
```

```r
data1
```

```r
head(data1)
```

```r
data1$age_group <- cut(data1$Age, c(-Inf, 0, 18, 24, 34, 44, 54, 64, Inf))
```

```r
summary(data1)
```

```r
install.packages('doBy')
```

```r
library('doBy')
```

```r
# age brackets
siterange <- function(x){c(length(x), min(x), mean(x), max(x))}
summaryBy(Age~age_group, data = data1, FUN=siterange)
```

```r
# only signed-in users have ages and genders
summaryBy(Gender+Signed_In+Impressions+Clicks~age_group, data = data1)
```

```r
# plot it!
install.packages("ggplot2")
library(ggplot2)
```

```r
#help(ggplot)
```

```r
#help(aes)
```

```r
ggplot(data = data1, mapping = aes(x=Impressions, fill=age_group))+geom_histogram(binwidth=1)
ggplot(data1, aes(x=age_group, y=Impressions, fill=age_group))+geom_boxplot()
```

__References:__

<a href="https://en.wikipedia.org/wiki/Click-through_rate">Definition: Click-through rate</a>
Ratio of number of users who view a page vs. Number of users who click a specific link

<a href="https://en.wikipedia.org/wiki/Impression_(online_media)">Definition: Impressions</a>
An impression is when an ad is fetched from its source and viewed in a user's browser. It doesn't matter if the ad is clicked.

```r
# analyze the click-through rate
# we want to get rid of clicks without impressions (i.e. when a user has an ad-blocker)
# otherwise our data will be skewed!

data1$hasImps <- cut(data1$Impressions, c(-Inf, 0, Inf))
summaryBy(Clicks~hasImps, data = data1, FUN = siterange)
ggplot(subset(data1, Impressions > 0), aes(x = Clicks/Impressions, color=age_group))+geom_density()
ggplot(subset(data1, Clicks > 0), aes(x = Clicks/Impressions, color=age_group))+geom_density()
ggplot(subset(data1, Clicks > 0), aes(x = age_group, y = Clicks, color = Clicks, fill = age_group))+geom_boxplot()
ggplot(subset(data1, Impressions > 0), aes(x = Clicks, color = age_group))+geom_density()
```

```r
# creating categories
data1$scode[data1$Impressions==0] <- "No Impressions"
data1$scode[data1$Impressions >0 ] <- "Impressions"
data1$scode[data1$Clicks > 0] <- "Clicks"
```

```r
#help(factor)
```

```r
# Conver the column to a factor
data1$scode <- factor(data1$scode)
head(data1)
```

```r
# look at levels
clen <- function(x){c(length(x))}
etable <- summaryBy(Impressions~scode+Gender+age_group, data = data1, FUN = clen)
```
