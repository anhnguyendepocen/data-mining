## Making Synthetic Datasets with sklearn.datasets

```python
# Generate data
from sklearn.datasets import make_blobs
import numpy as np
```

```python
X, y_ = make_blobs(n_samples=5000, centers=6, cluster_std=1.3, random_state=42)
labels = ["cyan", "yellow", "magenta"]
y = np.asarray([labels[int(i/2)] for i in y_])
```

```python
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style()
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(8,8))
for label in labels:
    plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.2)
```

## Importing / Loading External Data

We'll work with the Iris dataset.

```python
import pandas as pd
data = pd.read_csv('data/iris.csv')
```

Note:
* Each row is an instance
* Each column is a feature
* The last column contains the labels

```python
data.sample(10)
```

```python
from sklearn.model_selection import train_test_split
y = data.values[:, -1]
X = data.values[:,0:4]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## SKLearn's API has a consistent pattern...
To keep things simple and make it easy to swap models, all learning algos in SKLearn share a an API consisting of three interfaces:

- an `estimator`   for fitting models
- a  `predictor`   (you guessed it) for making predictions
- a  `transformer` for converting data

### Basic Code Pattern for Estimators

```python
class Estimator(object):
    def fit(self, X, y=None):
        """This fits the estimator to data."""
        # set the state of `self`
        return self
```

#### Example: k-NN

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=5)

clf.fit(X, y)
```

```python
# Estimator state is stored in instance attributes
clf._tree
```

### Basic Code Pattern for Predictors

```python
print(clf.predict(X)) 
```

```python
# Compute (approximate) class probabilities
print(clf.predict_proba(X))
```

## K-Nearest Neighbors

```python
# K-Nearest Neighbours
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

Model = KNeighborsClassifier(n_neighbors=8)
Model.fit(X_train, y_train)

y_pred = Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

###  Radius Neighbors Classifier
RNC implements a **vote** among neighbors within a given **radius**

**RadiusNeighborsClassifier** is very similar to **KNeighborsClassifier**.

```python
from sklearn.neighbors import RadiusNeighborsClassifier

Model=RadiusNeighborsClassifier(radius=1.)
Model.fit(X_train,y_train)

y_pred=Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

Wow. 100% accuracy on the test set.

## Logistic Regression
Logistic regression is appropriate to use when we are able to perform a 1-vs-all binary dichotomy between a given class and all other classes.

```python
# LogisticRegression
from sklearn.linear_model import LogisticRegression
Model = LogisticRegression(solver='lbfgs', multi_class='auto')
Model.fit(X_train, y_train)

y_pred = Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

###  Passive Aggressive Classifier

As per <a href="http://jmlr.csail.mit.edu/papers/volume7/crammer06a/crammer06a.pdf">this research paper from 2006</a>, the passive-aggressive classifier improves worst-case outputs for multiclass and linear classification problems.

**Note: This algorithm is very inconsistent given the size of our dataset!**

```python
from sklearn.linear_model import PassiveAggressiveClassifier
Model = PassiveAggressiveClassifier(max_iter=5000, tol=0.001)
Model.fit(X_train, y_train)

y_pred = Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

## Support Vector Machines

SVMs are perhaps most well-known for their effectiveness in high dimensional spaces. They remaim effective when there are more features than instances/samples. Furthermore, it tends to be relatively memory-efficient.

SVM's support kernel tricks that allow a data scientist to project a dataset into a different dimension in order to optimize the separation of classes.

## Other Ensembe Methods

### Bagging
>A Bagging classifier is an ensemble meta-estimator that fits base classifiers each on random subsets of the original dataset and then aggregate their individual predictions (either by voting or by averaging) to form a final prediction. Such a meta-estimator can typically be used as a way to reduce the variance of a black-box estimator (e.g., a decision tree), by introducing randomization into its construction procedure and then making an ensemble out of it.

>This algorithm encompasses several works from the literature. When random subsets of the dataset are drawn as random subsets of the samples, then this algorithm is known as Pasting. If samples are drawn with replacement, then the method is known as Bagging. When random subsets of the dataset are drawn as random subsets of the features, then the method is known as Random Subspaces. Finally, when base estimators are built on subsets of both samples and features, then the method is known as Random Patches.  

>from: [http://scikit-learn.org]

```python
from sklearn.ensemble import BaggingClassifier
Model=BaggingClassifier()
Model.fit(X_train,y_train)

y_pred=Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

###  AdaBoost classification

"An AdaBoost classifier is a meta-estimator that begins by fitting a classifier on the original dataset and then fits additional copies of the classifier on the same dataset but where the weights of incorrectly classified instances are adjusted such that subsequent classifiers focus more on difficult cases."

```python
from sklearn.ensemble import AdaBoostClassifier
Model=AdaBoostClassifier()
Model.fit(X_train,y_train)

y_pred=Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

### Gradient Boosting
Forward stage-wise fashion for building a model; it allows optimization of an arbitrary loss function (as long as the function is differentiable!).

```python
from sklearn.ensemble import GradientBoostingClassifier
Model=GradientBoostingClassifier()
Model.fit(X_train,y_train)

y_pred=Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

### LDA - Linear Discriminant Analysis

Cloely related to ANOVA and regression analysis.
LDA and Quadratic Discriminant Analysis are *classic* techniques.

They have closed-form solutions that are generally easy to compute, are multiclass by nature, work well in practice, and.... get this: have no **hyperparameters** to tune.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
Model=LinearDiscriminantAnalysis()
Model.fit(X_train,y_train)

y_pred=Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

### QDA - Quadratic Discriminant Analysis
The model fits a Gaussian density to each class. Classifier with a quadratic decision boundary. Generated by fitting class-dependent densities to the data using Bayes’ rule to vote.

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
Model=QuadraticDiscriminantAnalysis()
Model.fit(X_train,y_train)

y_pred=Model.predict(X_test)

# Summarize predictions, print metrics
print('CLASSIFICATION REPORT:\n', classification_report(y_test, y_pred))
print('\n CONFUSION MATRIX:\n',   confusion_matrix(y_test, y_pred))
print('\n\n ACCURACY: ',          accuracy_score(y_pred, y_test))
```

## K-means Clustering
K-means clustering is a type of unsupervised learning, which is used when you have unlabeled data (i.e., data without defined categories or groups). 

This algorithm is designed **to find groups in the data**. The number of groups is represented by the variable K.

The algorithm assigns each point to a class K based on its features.
