# Centroid of a point cloud

```python
# Generate data
from sklearn.datasets import make_blobs
import numpy as np
import pandas as pd

# plotting
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style()
%matplotlib inline
```

## Making synthetic data

```python
# produce 2 clusters, each with a differend standard dev
X_, y_ = make_blobs(n_samples=1000, centers=2, cluster_std=[1.6, 0.8], center_box = (-3, 8), random_state=42)
labels = ["cyan", "magenta"]
y = np.asarray([labels[int(i)] for i in y_])
```

```python
# transform the data 
transform = [[0.60834549, -0.63667341], [-0.10887718, 0.85253229]]
X = np.dot(X_, transform)
```

```python
# turn the data into a dataframe, particularly for the convenience of concatenating arrays
data_ = pd.DataFrame(X, columns=['x1', 'x2'])
data_['labels'] = y

# sort the data by class label
data_ = data_.sort_values(by=['labels'])

# show sample of the data
display(data_.sample(10))

# check to see how many of each class
display(data_['labels'].value_counts())
```

```python
# drop some randomly selected magenta instances to simulate eneven dataset
data_ = data_.drop(data_.index[np.random.randint(501, 1000, 1000)])

data_['labels'].value_counts()
```

```python
# convert to numpy array
data = data_.values
```

```python
data[501:, 1] = data[501:, 1]*4 + 5.5
```

## Finding centroids and plotting

```python
fig, ax = plt.subplots(figsize=(9,9))
ax.grid(False)

# plot the point clouds
for label in labels:
    plt.scatter(data[:, 0], data[:, 1], c=data[:, 2], alpha=0.2)

# plot the centroids of each cloud
    # calculate the centroid of each cluster
clust1 = data[0:500, :]
clust2 = data[500:1000, :]

    # centroid coordinates for clust1
cx1 = np.mean(clust1[:,0])
cy1 = np.mean(clust1[:,1])

    # centroid coordinates for clust2
cx2 = np.mean(clust2[:,0])
cy2 = np.mean(clust2[:,1])

    # plot centroids
plt.scatter([cx1,cx2],[cy1,cy2], c='r', edgecolors='w', linewidth='3', marker='o', s=100)

    # line connecting centroids
plt.plot([cx1,cx2],[cy1,cy2], c='r', linewidth='3', alpha=1)

    # plot line through centroids
u = np.linspace(-3, 4, 100)
m = (cy2-cy1)/(cx2-cx1)  # y = mx + b ..... v = mu + b
v = m*u + (cy2 - m*cx2)  # y = mx + b ..... v = mu + b
plt.plot(u, v)
```
