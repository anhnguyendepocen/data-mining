# A Comprehensive Guide to *Git*

## What is Git?

Git is an *Open Source Distributed Version Control System*.<br /><br />

But what does that mean? In a nutshell: <br /><br />

Git stores content and code, allowing one *or more* developers to contribute to a project in parallel while and efficiently managing changes throughout development. It allows a developer or development team to control the release of new and changing versions of a codebase. <br /><br />

Git maintains a history of changes. <br /><br />

The fact that Git is a *distributed* content tracking system means that a full version of the code isn't just present in a remote server, but also present in each developer's machine. So Git allows for both remote and local repositories. <br /><br />

All of this allows developers to work in parallel, manage conflicts as parallel branches merge back into the codebase, and even revert to previous versions of the code when necessary. 

## Why will we use Git?

First of all, out in the real world most of the development that happens is by teams of people working together.

Second of all, this is a way individual developers to make their code available to themselves across different machines and to make their work into a well-organized public resource and career profile.

Mastering a version control system, Git in particular, is essential for establishing a career and working as well as possible with teams and on individual projects.

The `git init` command adds a local Git repository at the location in which the command is performed.

## Staging and Committing Code

Committing is the process of adding code to a local repository. In order for code to be committed, it must be staged in the staging area. The job of the staging area is to track all of the code that is intended to be committed; if something isn't in the staging area, it wont be committed to the local repository. This gives you (the developer) control over which files will be committed. Pretty snazzy, huh?

### Staging your code

Let's stage the file we just added to our repository folder. Use the following command in your terminal:

```

git add '1 - Satistics in Data Science.ipynb'


```

As an aside, if you want to add multiple files at the same time you use:

`git add file1 file2 file3`

And to add *all* the files in your project folder to the staging area, use:

`git add .`

but be careful. Remember that it adds *all* of the files.

### Committing

Use the following command to commit the file. When you commit you should use a relevant message in order to indicate what changes were performed in this commit.

```

git commit -m "Inital Commit"


```
