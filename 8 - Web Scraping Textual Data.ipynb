```python
import logging
import scrapy
from scrapy.crawler import CrawlerProcess

class ArticleSpider(scrapy.Spider):
    name = "article"
    start_urls = ['https://badwitch.es/3-ways-dance-liberated-me/',
                 'https://badwitch.es/the-shadow-powerful-methods-for-shadow-integration/']
    
    custom_settings = { 
        'LOG_LEVEL': logging.WARNING, 
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1},  # Used for pipeline 1
        'FEED_FORMAT':'csv',                                  # Used for pipeline 2
        'FEED_URI': 'article.csv'                             # Used for pipeline 2
     }    
    
    def parse(self, response):
        for article in response.css("div.entry-content"):
            content = response.xpath(".//div[@class='entry-content']/descendant::text()").extract()
            yield {'article': ' '.join(content)}
```

```python
import json
 
class JsonWriterPipeline(object):
    def open_spider(self, spider): 
        self.file = open('article2.jl', 'w')
    def close_spider(self, spider): 
        self.file.close()
    def process_item(self, item, spider): 
        line = json.dumps(dict(item)) + "" 
        self.file.write(line)
        return item
```

```python
import logging
import scrapy
from scrapy.crawler import CrawlerProcess

class BlogSpider(scrapy.Spider):
    name = "article"
    start_urls = ['https://badwitch.es/']
    
    def parse(self,response):
        #cat_list = []
        for category_url in  response.xpath(".//div[@id='df-side-menu']//li/descendant::a/@href").extract():
            yield response.follow(category_url, callback=self.parse_page_urls)
            #cat_list.append(category_url)
        #print(cat_list)
    
    def parse_page_urls(self,response):
        
        #article_list = []
        for article_url in response.xpath(".//h4[@class='article-title']/a//@href").extract():
            yield response.follow(article_url, callback=self.parse_article)
            #article_list.append(article_url)
        
        next_page = response.xpath(".//a[@class='next page-numbers']/@href").extract_first()
        
        #page_list = []
        if next_page is not None:
            yield response.follow(next_page, callback=self.parse_page_urls)
            #page_list.append(next_page)
        
        #print(page_list)
        #print(article_list)
    
    custom_settings = { 
        'LOG_LEVEL': logging.WARNING, 
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1},  # Used for pipeline 1
        'FEED_FORMAT':'csv',                                   # Used for pipeline 2
        'FEED_URI': 'article2.csv'                              # Used for pipeline 2
     }    
    
    def parse_article(self, response):
        #for article in response.css("div.entry-content"):
        
        content_1 = response.xpath(".//div[@class='entry-content']/*[self::p or self::h5 or self::h4 or self::h3 or self::h2 or self::h1][not(descendant::img)][not(preceding::p[contains(., 'IN CONCLUSION')])][not(descendant-or-self::p[contains(., 'IN CONCLUSION')])][not(preceding::p[contains(., 'About the Author:')])][not(descendant-or-self::p[contains(., 'About the Author')])][not(preceding::p[contains(., 'Get the weird key to making magic work')])][not(descendant-or-self::p[contains(., 'Get the weird key to making magic work')])]").extract()
        content_2 = response.xpath(".//div[@class='entry-content']/descendant::text()[not(descendant::img)][not(preceding::text()[contains(., 'IN CONCLUSION')])][not(descendant-or-self::text()[contains(., 'IN CONCLUSION')])][not(preceding::text()[contains(., 'About the Author:')])][not(descendant-or-self::text()[contains(., 'About the Author')])][not(preceding::text()[contains(., 'Get the weird key to making magic work')])][not(descendant-or-self::text()[contains(., 'Get the weird key to making magic work')])][not(descendant-or-self::text()[contains(., 'image source')])]").extract()
        
        title = response.xpath(".//h1[@class='entry-title']/text()").extract()
        author = response.xpath(".//div[@class='authors-meta']//a//text()").extract_first()
        post_date = response.xpath(".//div[@class='authors-meta']//span//text()[contains(., ',')]").extract_first()
        img_urls = ""
        tags = response.xpath(".//ul[@class='list-inline tags']//a[@rel='tag']/text()").extract()
        
        views = response.xpath(".//i[@class='ion-eye']/following::text()[contains(., 'views')]").extract_first()
        try: views = views[:-6]
        except: views = ''
        
        shares = response.xpath(".//div[@class='style1 df-post-sharing  df-social-share']//li//descendant::h3/text()").extract_first()
        
        #extract page url from response variable
        url = list((str(response).split("200")[1]))
        try: del url[-1]
        except: pass
        try: del url[0]
        except: pass
        #print(''.join(url))
        
        yield {
            'article_v1': ' '.join(content_1),
            'article_v2': ' '.join(content_2),
            'title': title,
            'author': author,
            'post_date': post_date,
            'shares': shares,
            'views': views,
            'img_urls': img_urls,
            'tags': tags,
            'url': ''.join(url)            
        }
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(BlogSpider)
process.start()
```

##### Clean it

```python
import json
import pandas as pd
import datetime
import os

cwd = os.getcwd()+"\\article2.csv"
path = r''+cwd
df = pd.read_csv(path)
```

```python
for i in range(len(df)):
    splitter = df.iloc[i][1]
    try: 
        splitter = ''.join(str(df.iloc[i][1]).split("\r\n")[1:2]) # select author name
        splitter = ''.join(str(df.iloc[i][1]).split(splitter))    # remove author name
    except: pass

    df.iloc[i, 1] = splitter

# make empty list that will become additional column on df
# make empty list that will tell us which indices to skip for author removal
writers = []
    
for i in range(len(df)):
    skip = 0
    
    # get writer's name, append to author list
    writer = ''.join(a for a in str(df.iloc[i][0]).split("<p>")[1:2])
    writer = ''.join(a for a in writer.split("</p>"))
    
    if "by " in writer: writer = ''.join(writer.split("by "))
    else: writer = ''
    
    if "<" in writer or ">" in writer: 
        writer = ''
        skip = 1
    
    if len(writer) > 50: writer = ''
    
    writers.append(writer)
    
    if skip == 0:
        try: 
            splitter = ''.join(str(df.iloc[i][0]).split("</p>")[0:1]) # select author name
            splitter = ''.join(str(df.iloc[i][0]).split(splitter))    # remove author name
            splitter = ''.join(str(splitter).split("</p>"))           # remove trailing </p> tag
        except: pass
        df.iloc[i, 0] = splitter
    
if len(writers) == len(df):
    try:
        df['writers'] = writers
    except: print("error")

df['writers']
```

```python
import logging
import scrapy
from scrapy.crawler import CrawlerProcess

class BlogSpider(scrapy.Spider):
    name = "article"
    start_urls = ['https://www.yourtango.com/love',
                 #'https://www.yourtango.com/zodiac',
                 ]
    
    def parse(self,response):
        
        #article_list = []
        for article_url in response.xpath(".//div[@class='teaser-mix-left']/descendant::a//@href").extract():
            article_url = "https://yourtango.com/"+article_url
            yield response.follow(article_url, callback=self.parse_article)
            #article_list.append(article_url)
        
        next_page = response.xpath(".//li[@class='next']/descendant::a/@href").extract_first()
        
        #page_list = []
        if next_page is not None:
            next_page = "https://yourtango.com/"+next_page
            yield response.follow(next_page, callback=self.parse)
            #page_list.append(next_page)
        
        #print(page_list)
        #print(article_list)
    
    custom_settings = { 
        'LOG_LEVEL': logging.WARNING, 
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1},  # Used for pipeline 1
        'FEED_FORMAT':'csv',                                   # Used for pipeline 2
        'FEED_URI': proj_name+'.csv'                           # Used for pipeline 2
     }    
    
    def parse_article(self, response):
        #for article in response.css("div.entry-content"):
        
        # extract 2 versions of each article to accomodate for different xpath structures
        content_1 = response.xpath(".//div[@class='yt-node-content']/descendant::p/text()[not(self::*[@class='article-body-related-links'])][not(descendant::iframe)][not(descendant::a/@href[contains(., 'giphy')])][not(descendant::a/text()[contains(., 'RELATED:')])] |.//div[@class='yt-node-content']/descendant::a/text()[not(ancestor::*[@class='article-body-related-links'])][not(ancestor::article)][not(descendant-or-self::text()[contains(., 'GIPHY')])][not(descendant-or-self::text()[contains(., 'RELATED')])]").extract()
        content_2 = response.xpath(".//div[@class='yt-node-content']/descendant::p[not(self::*[@class='article-body-related-links'])][not(descendant::iframe)][not(descendant::a/@href[contains(., 'giphy')])][not(descendant::a/text()[contains(., 'RELATED:')])][not(descendant::*[@class='share_pinterest_wrapper'])] | .//div[@class='yt-node-content']/descendant::h2[not(self::*[@class='article-body-related-links'])][not(descendant::iframe)][not(descendant::a/@href[contains(., 'giphy')])][not(descendant::a/text()[contains(., 'RELATED:')])][not(ancestor::*[@class='article-body-related-links'])][not(ancestor::article)][not(descendant-or-self::text()[contains(., 'GIPHY')])][not(descendant-or-self::text()[contains(., 'RELATED')])]").extract()
        post_date = response.xpath(".//div[@class='post-date']/text()").extract_first()
        author = response.xpath(".//div[@class='author-name']//a//text()").extract_first()
        categories = response.xpath(".//div[@class='author-name']//a[@class='category']//text()").extract()
        title = response.xpath(".//h1[@class='page-header']/text()").extract()
        img_urls = response.xpath(".//div[@class='yt-node-content']/descendant::img/@src").extract()
        
        #extract page url from response variable
        url = list((str(response).split("200")[1]))
        try: del url[-1]
        except: pass
        try: del url[0]
        except: pass
        #print(''.join(url))
        
        yield {
            'article_v1': ' '.join(content_1),
            'article_v2': ' '.join(content_2),
            'title': title,
            'author': author,
            'post_date': post_date,
            'img_urls': img_urls,
            'categories': categories,
            'url': ''.join(url)            
        }
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(BlogSpider)
process.start()
```

#### Putting it all together

```python
from bs4 import BeautifulSoup
from stop_words import get_stop_words
from nltk.corpus import stopwords
import nltk
import pandas as pd


def wordFreqDist(input_string, excl = []):
    try:
        soup = BeautifulSoup(input_string)
        text = soup.get_text(strip=True)
        tokens = [t for t in text.split()]
        freq = nltk.FreqDist(tokens)

        stop_words = list(get_stop_words('en'))         #About 900 stopwords
        nltk_words = list(stopwords.words('english'))   #About 150 stopwords
        stop_words.extend(nltk_words)
        stop_words.extend(excl)

        #removes stopwords and non-alphanumeric characters
        clean_tokens = ([token for token in tokens if token not in stop_words])
        freq = nltk.FreqDist(clean_tokens)
        fig = freq.plot(20, cumulative=False)
        return fig
    except: print("error")
```

## Textblob: Quick Sentiment Analysis

```python
from textblob import TextBlob
from textblob import Blobber
from textblob.sentiments import NaiveBayesAnalyzer
from textblob.sentiments import PatternAnalyzer
```

```python
for title in df['title'][0:5]:
    print(title)
    print(TextBlob(title).sentiment.polarity)
```

```python
tb_nba = Blobber(analyzer=NaiveBayesAnalyzer())
tb_pa = Blobber(analyzer=PatternAnalyzer())

for title in df['title'][0:5]:
    print(title)
    print(tb_nba(title).sentiment)
    print(tb_pa(title).sentiment)
```

```python
title_polarity = [round(float(tb_pa(title).sentiment.polarity), 3) for title in df['title']]
df_2 = pd.DataFrame([title_polarity, list(df['title'])]).T
```

```python
df_2.columns = ['sentiment_score', 'title']
```
