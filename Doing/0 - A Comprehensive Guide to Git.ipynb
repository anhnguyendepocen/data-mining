# A Comprehensive Guide to *Git*

## What is Git?

Git is an *Open Source Distributed Version Control System*.<br /><br />

But what does that mean? In a nutshell: <br /><br />

Git stores content and code, allowing one *or more* developers to contribute to a project in parallel while and efficiently managing changes throughout development. It allows a developer or development team to control the release of new and changing versions of a codebase. <br /><br />

Git maintains a history of changes. <br /><br />

The fact that Git is a *distributed* content tracking system means that a full version of the code isn't just present in a remote server, but also present in each developer's machine. So Git allows for both remote and local repositories. <br /><br />

All of this allows developers to work in parallel, manage conflicts as parallel branches merge back into the codebase, and even revert to previous versions of the code when necessary. 

## Why will we use Git?

First of all, out in the real world most of the development that happens is by teams of people working together.

Second of all, this is a way individual developers to make their code available to themselves across different machines and to make their work into a well-organized public resource and career profile.

Mastering a version control system, Git in particular, is essential for establishing a career and working as well as possible with teams and on individual projects.

## Let's get started:

### Download and Install Git

Go to this link and choose the install that is appropriate for your operating system:
https://git-scm.com/book/en/v2/Getting-Started-Installing-Git

Verify if Git is installed by using a terminal / command prompt window to run the following command:

```

git --version


```

### Create a new local Git repository

Create a new folder in your computer where you'd like to keep your repositories. Mine is called `data-science-and-research-skills` since that is the project I am working on.

Now, in your terminal, travel to that location and initialize git. The machine I am currently using is Windows, so for me this looks like:

```
cd c:\Users\MyUsername\Documents\my-git-repositories\data-science-and-research-skills
git init

```

The `git init` command adds a local Git repository at the location in which the command is performed.

### Add some code to your repository

Since this is part of a series for building important fundamental data science and research skills, we'll make a notebook for our next project, which is *Statistics in Data Science*. Go ahead and create a new notebook file. ALternatively, add whatever type of file you would like. Mine is going to be called 

`1 - Satistics in Data Science.ipynb`

I like to follow my own file naming convention, which makes it easier to automate the process of turning my notebooks into blog posts and other publication-ready forms.

## Staging and Committing Code

Committing is the process of adding code to a local repository. In order for code to be committed, it must be staged in the staging area. The job of the staging area is to track all of the code that is intended to be committed; if something isn't in the staging area, it wont be committed to the local repository. This gives you (the developer) control over which files will be committed. Pretty snazzy, huh?

### Staging your code

Let's stage the file we just added to our repository folder. Use the following command in your terminal:

```

git add '1 - Satistics in Data Science.ipynb'


```

As an aside, if you want to add multiple files at the same time you use:

`git add file1 file2 file3`

And to add *all* the files in your project folder to the staging area, use:

`git add .`

but be careful. Remember that it adds *all* of the files.

### Committing

Use the following command to commit the file. When you commit you should use a relevant message in order to indicate what changes were performed in this commit.

```

git commit -m "Inital Commit"


```

## Git Status and Git Log

Now we are going to create the first change in our file. For my notebook, I will simply add a new markdown cell containing the title of the notebook. Let's keep it simple.

### Status

In order to see relevant information regarding which files have been modified and which are in the staging area (as well as other information that we aren't immediately concerned with), use the command 

```

git status


```

When you do this you'll notice that Git returns a message telling you that your file has been changed, but is not in the staging area. Add the changed file to the staging area and commit the file using the commands we already know. Be sure to add a relevant comment:

```

git add "1 - Statistics in Data Science.ipynb"
git commit -m "first changed to the Stats in Data Science guide"


```

### Log

If you want to see a history of all commits that have been done up to this point, use the command 

```

git log


```

As you can see, the log file returns information about each commit including the the commit id, author, date, and comment, as well as some other information.

## Branches

We have been working directly in the __master__ branch, which is the default option. We can create other branches, though. But what is a branch?

*A branch is simply a pointer to the latest commit in the Git repository,* and currently our __master__ branch is a pointer to the most recent commit.

### Why are multiple branches necessary?

In a nutshell, on development teams (or in individual work), there are multiple features of a project that are being developed in parallel. For instance, a team that develops a social media platform might have one team that develops their messaging system, and another that develops an event-creation feature. They both operate on the same code base, the code for the platform itself. They have to be able to work simultaneously, commit changes, and resolve any conflicting changes in code. This is where branches and merges come into play. We'll talk about merges shortly.

When a new branch is created, it diverges from the __master__ branch. When it is ready to be added to the master branch, it is merged back into the master codebase.

### Create a new Local Branch

Create a new branch called __test__:

```

git branch test


```

We are still working int he context of the __master__ branch, so lets switch to the __test__ branch:

```

git checkout test


```

We're in the test branch, as your Git terminal should tell you. We can list all of the branches with

```

git branch


```

When you do so, it should indicate with an asterisk the branch that you are currently in.

### Try committing to the New Branch

Modify your file. In my case, I will add some text to my notebook `1 - Statistics in Data Science.ipynb`.

Then stage and commit using the commands you already learned. In my case, I will use the following:

```

git add "1 - Statistics in Data Science.ipynb"
git commit -m "Test branch commit - learning how to create and merge branches"


```

Now the __test__ branch is one commit ahead of the __master__ branch since the __test__ branch already includes the two commits from __master__. Run `git log` to verify for yourself. You should see the commits listed in descending order.

### Merges

Now perhaps we want to `merge` the __test__ branch back into the __master__ branch. This is where we use `git merge`.

To do this, use the following commands step by step:

Switch to the __master__ branch: `git checkout master`
Run the `merge` command on the __test__: `git merge test`

Since there are no conflicts in this simple example, the merge should be successful. In more complex situations (real life projects) there will be conflicts. Resolving those conflicts will come with experience. On developments teams there will always be a lead developer who can (hopefully) resolve these conflicts as you learn to do it yourself. Until then, strive to learn how to resolve them before the code gets to your lead developer.

Go ahead and run `git log` within the __master__ branch. You should see all three commits.

## Remote Git Repositories

We've been working in a local repository, but in the real world we as developers push and pull data to and from remote repositories. Once our cade is in a remote repository, and developer who can also access that remote repository can see and modify that code.

### GitHub

GitHub is the remote repository I am used to using. It serves as my portfolio and a way to work on projects with others and access my projects from different machines.

If you are following along and don't already have a GitHub account, head over to https://github.com/ and create an account.

### Create a remote repository

Once you have an account, go ahead and make a new repository on GitHub.com. Using the GitHub website should make this relatively easy.

I am going to name my github repository `data-science-and-research-skills`, which is the same name as my local repository. They do not have to have the same name.

Your repository on GitHub will be given its own URL. Mine has the URL `https://github.com/clockelliptic/data-science-and-research-skills.git`

Your repository will have a similar URL `https://github.com/clockelliptic/YOUR-REPOSITORY-NAME.git`

### Point a local repository toward a remote repository

Now that you have your local repository and the remote repository, you want to prepare your local repository so that you can `push` your local code to the remote codebase.

Point the local repository to the remote repository with:

```

git remote add origin https://github.com/clockelliptic/data-science-and-research-skills.git


```

*Be sure to use your own repository URL when you do this.*

### Git Push

Now we want to push all of the local code to the remote repository. To do this, use the following:

```

git push -u origin master


```

This pushes all content in our local master branch to the master branch of the remote repository.

## More Commands

### Git Pull
 `git pull` is used to pull new changes in the remote repository into your local repository. This is necessary to keep your local code up to date as other developers continue to update the remote repository with their work. To update your local repository use
 
 ```
 
 git pull origin master
 
 
 ```
 
If you have access to two machines, have the time to create a Virtual Machine, or venture into using WSL or other comparable options, you can practice doing this yourself. But first, you might want to know how to `clone` a repository...

### Git Clone

`git clone` is used to create a copy of an existing repository on your own local machine. To do this, use this command:

```

git clone [url of repository]


```
